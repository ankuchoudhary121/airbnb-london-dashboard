```python
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Dropping NA values
#df = df.drop(['neighbourhood_group_cleansed', 'calendar_updated', 'license'], axis=1)

df = df.drop([
    'listing_url', 'picture_url', 'host_thumbnail_url', 'host_picture_url',   # URLs
    'description', 'neighborhood_overview', 'host_about',                     # Long text
    'host_verifications',                                                     # List column
    'calendar_last_scraped', 'last_scraped',                                  # Metadata
    'maximum_minimum_nights', 'minimum_maximum_nights',
    'maximum_maximum_nights', 'minimum_nights_avg_ntm', 'maximum_nights_avg_ntm'  # Rarely used constraints
], axis=1)

```

```python
#Numeric columns — FILL with median

num_cols = ['bathrooms','bedrooms','beds',
            'review_scores_accuracy','review_scores_cleanliness',
            'review_scores_checkin','review_scores_communication',
            'review_scores_location','review_scores_value',
            'review_scores_rating','reviews_per_month']

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

```python
df.columns[df.isnull().any()]
```

```python
df = df.drop([
    'host_listings_count',
    'host_total_listings_count',
    'host_has_profile_pic',
    'host_identity_verified',
    'minimum_minimum_nights'
], axis=1)
```

```python
cat_cols = [
    'host_name','host_since','host_location',
    'host_response_time','host_response_rate','host_acceptance_rate',
    'host_is_superhost','host_neighbourhood','bathrooms_text',
    'has_availability'
]

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.columns[df.isnull().any()]
```

```python
df['price'] = df['price'].fillna(df['price'].median())
df['neighbourhood'] = df['neighbourhood'].fillna(df['neighbourhood'].mode()[0])
```

```python
df.columns[df.isnull().any()]
```

```python
# Drop Duplicates
df = df.drop_duplicates()
```

```python
# filtering unrealistic data
df = df[(df['price'] > 0) & (df['price'] < 1000)]
```

```python
#Feature Engineering (Create New Columns)
#1.Price category: budget / mid / luxury

def price_category(p):
    if p < 50:
        return "Budget"
    elif p < 150:
        return "Mid-range"
    else:
        return "Luxury"

df['price_category'] = df['price'].apply(price_category)



```

```python
# 2.Host type: single-listing vs multi-listing host

def host_type(count):
    if count == 1:
        return "Single-listing host"
    elif count <= 5:
        return "Small host"
    else:
        return "Professional host"

df['host_type'] = df['calculated_host_listings_count'].apply(host_type)
```

```python
# Price distribution – Histogram
plt.figure()
sns.histplot(df['price'], bins=50)
plt.xlabel("Price per night (£)")
plt.title("Distribution of Airbnb prices in London")
plt.show()
```

```python
# Average price by room type – Bar plot
avg_price_room = df.groupby('room_type')['price'].mean().reset_index()

plt.figure()
sns.barplot(data=avg_price_room, x='room_type', y='price')
plt.title("Average price by room type")
plt.ylabel("Average price (£)")
plt.show()
```

```python
avg_price_neigh = (
    df.groupby('neighbourhood_cleansed')['price']
      .mean()
      .sort_values(ascending=False)
      .head(10)
      .reset_index()
)

plt.figure()
sns.barplot(data=avg_price_neigh, x='price', y='neighbourhood_cleansed')
plt.title("Top 10 neighbourhoods by average price")
plt.xlabel("Average price (£)")
plt.show()
```

```python
plt.figure()
sns.boxplot(data=df, x='room_type', y='price')
plt.ylim(0, 500)  # to ignore extreme outliers visually
plt.title("Price distribution by room type")
plt.show()
```

```python
plt.figure()
sns.scatterplot(data=df, x='number_of_reviews', y='price', alpha=0.3)
plt.xlim(0, 200)  # limit to make it readable
plt.title("Price vs Number of Reviews")
plt.show()
```

```python
# Heatmap
num_cols = ['price', 'minimum_nights', 'number_of_reviews',
            'availability_365', 'reviews_per_month']

corr = df[num_cols].corr()

plt.figure()
sns.heatmap(corr, annot=True)
plt.title("Correlation between key numerical variables")
plt.show()
```

```python
#demand index: high reviews and low availability → high demand.
def demand_index(row):
    if (row['number_of_reviews'] > 50) and (row['availability_365'] < 100):
        return "High demand"
    elif (row['number_of_reviews'] < 10) and (row['availability_365'] > 200):
        return "Low demand"
    else:
        return "Medium demand"

df['demand_index'] = df.apply(demand_index, axis=1)
```

```python
query3 = """
SELECT neighbourhood_cleansed, AVG(price) AS avg_price
FROM listings
WHERE room_type = 'Private room'
GROUP BY neighbourhood
ORDER BY avg_price DESC
LIMIT 10;
"""

pd.read_sql_query(query3, conn)
```
